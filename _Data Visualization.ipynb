```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset.
np.random.seed(0)
x = np.random.rand(100)
y = x * 0.5 + np.random.normal(0, 0.1, 100)
df1 = pd.DataFrame({'X': x, 'Y': y})
sns.scatterplot(data=df1, x='X', y='Y')
plt.title('Scatter Plot of X vs Y')
plt.show()
```

```python
# 2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
data = np.random.randn(1000)
sns.histplot(data, kde=True)
plt.title('Distribution of Random Numbers')
plt.show()
```

```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
categories = ['A', 'B', 'C', 'D']
values = [np.random.randint(1, 10) for _ in categories]
df2 = pd.DataFrame({'Category': categories, 'Value': values})
sns.barplot(data=df2, x='Category', y='Value')
plt.title('Comparison of Categories')
plt.show()
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
np.random.seed(1)
categories = np.random.choice(['Cat1', 'Cat2', 'Cat3'], size=200)
values = np.random.randn(200)
df3 = pd.DataFrame({'Category': categories, 'Value': values})
sns.boxplot(data=df3, x='Category', y='Value')
plt.title('Distribution Across Categories')
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
np.random.seed(2)
x = np.random.rand(100)
y = x * 0.7 + np.random.normal(0, 0.2, 100)
z = np.random.rand(100)
df4 = pd.DataFrame({'X': x, 'Y': y, 'Z': z})
corr = df4.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(x+y)

import numpy as np
import pandas as pd
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=df['Z'].values.reshape(100, 100),
                                x=df['X'].values.reshape(100, 100),
                                y=df['Y'].values.reshape(100, 100))])
fig.update_layout(title='3D Surface Plot', scene=dict(zaxis=dict(range=[-1, 1])))
fig.show()
```

<P3><b>BOKEH ASSIGNMENT</b>

```python
#1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2)

show(p)
```

```python
#2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import linear_cmap
from bokeh.io import show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.palettes import Viridis256

output_notebook()

x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 20
colors = np.random.rand(50)

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))
color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')
p.scatter('x', 'y', size='sizes', color=color_mapper, source=source, fill_alpha=0.6)

show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

from bokeh.plotting import figure, show, output_notebook

output_notebook()

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(x_range=fruits, height=250, width=400, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x=fruits, top=counts, width=0.9)

p.xgrid.grid_line_color = None
p.y_range.start = 0
p.axis.minor_tick_line_color = None
p.outline_line_color = None

show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.

import numpy as np
from bokeh.plotting import figure, show, output_notebook

output_notebook()

data_hist = np.random.randn(1608)
hist, edges = np.histogram(data_hist, bins=30)

p_hist = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency', tools="")
p_hist.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

show(p_hist)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.

import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import LinearColorMapper
from bokeh.palettes import Viridis256
from bokeh.transform import linear_cmap

output_notebook()

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

data = {'x': xx.flatten(), 'y': yy.flatten(), 'z': data_heatmap.flatten()}

p_heatmap = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', x_range=(0, 1), y_range=(0, 1), tools="", toolbar_location=None)
color_mapper = LinearColorMapper(palette=Viridis256, low=data['z'].min(), high=data['z'].max())
p_heatmap.rect(x='x', y='y', width=0.1, height=0.1, source=data, line_color=None, fill_color={'field': 'z', 'transform': color_mapper})

p_heatmap.xaxis.axis_label = None
p_heatmap.yaxis.axis_label = None
p_heatmap.xaxis.visible = False
p_heatmap.yaxis.visible = False

show(p_heatmap)
```
